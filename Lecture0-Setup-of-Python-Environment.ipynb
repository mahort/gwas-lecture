# Setup of Python Environment for GWAS-Lectures

We will be using [limix](https://github.com/limix/limix) for the GWAS-lectures.   
Limix is a flexible and efficient linear mixed-model library with interfaces to Python.  

Assuming that Python is installed on the system the following sections describe how to install limix and all it's dependencies. 

## Install conda

We will be using the open source package manager [conda](https://conda.io/docs/) to intall limix and jupyter notebook in a conda environment. Conda allows users to manage multiple environments with different package versions and can not only install python libraries but also various non-python software packages.  

There are two ways to install conda:
 * [Anaconda](https://www.anaconda.com/) (kitchensink containing optimized scientific R and python libraries)
 * [Miniconda](https://conda.io/miniconda.html) (minimal installer for conda)
 
We will use miniconda to install conda on our laptops. Navigate to https://conda.io/miniconda.html and download the version for your Operating System.

### Install conda on MacOS

```bash
curl https://repo.anaconda.com/miniconda/Miniconda3-latest-MacOSX-x86_64.sh --output Miniconda3-latest-MacOSX-x86_64.sh
bash Miniconda3-latest-MacOSX-x86_64.sh
```

### Install conda on Linux

```bash
curl https://repo.continuum.io/miniconda/Miniconda3-latest-Linux-x86_64.sh -O Miniconda3-latest-Linux-x86_64.sh
sh Miniconda3-latest-Linux-x86_64.sh
```

### Install conda on Windows (WSL)

The easiest way is to first install the [Windwos Subsystem for Linux (WSL)](https://docs.microsoft.com/en-us/windows/wsl/install-win10)
Once this is done start the Linux distrubition that was previously installed from the Start Menu

```bash
curl https://repo.continuum.io/miniconda/Miniconda3-latest-Linux-x86_64.sh -O Miniconda3-latest-Linux-x86_64.sh
sh Miniconda3-latest-Linux-x86_64.sh
```


## Create a conda environment and install limix, jupyter notebook

### For Linux, Mac OSX and WSL follow this:

```bash
$ git clone https://github.com/timeu/gwas-lecture.git 
$ # or download archive from and extract it from https://github.com/timeu/gwas-lecture/archive/master.zip
$ cd gwas-lecture
$ conda env create -f environment.yml
```


## Docker container

As a fallback if any of the installations above fail, it is possible to use a docker container that comes with all libraries pre-installed. First make sure to install [Docker](https://docs.docker.com/install/) on your machine.
Once docker is installed just run: 

```
$ docker run -it -p 8888:8888 timeu/gwas-lecture
```
Open http://localhost:8888 in your browser and copy and paste the token for the docker run above


## Start jupyter notebook

```
$ conda activate gwas-lecture
$ jupyter notebook
```
Open link that is displayed in terminal in browser
